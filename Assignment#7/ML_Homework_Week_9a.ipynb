```python
def beautify_txt(text):
    final_data = (((text).replace(u'\xa0', u' ')).replace(r'\r',u'\n').replace(r'\n', ' '))
    return final_data

Sylvia_Plath_Poems=["https://allpoetry.com/Balloons",
                    "https://allpoetry.com/A-Winter-Ship"]

Rabindranath_Tagore_Poems=[
                "https://allpoetry.com/Dream-Girl---------English-Translation",
                "https://allpoetry.com/At-The-Last-Watch"]    
poems_results_SP=[]
poems_results_RT=[]

def scrapeSPPoetsPoems():
  for url in Sylvia_Plath_Poems:
    header = {'User-Agent':'Mozilla/5.0'}
    requests = urllib.request.Request(url,headers=header)
    requestread = urllib.request.urlopen(requests).read()
    soup = bs.BeautifulSoup(requestread,'html.parser')
    poem_content = (beautify_txt(soup.find_all('div', class_="poem_body")[0].text))
    poem_title = soup.find_all('h1')[0].text
    poems_results_SP.append([poem_title, poem_content])
  return poems_results_SP

def scrapeRTPoetsPoems():
  for url in Rabindranath_Tagore_Poems:
    print(url)
    header = {'User-Agent':'Mozilla/5.0'}
    requests = urllib.request.Request(url,headers=header)
    requestread = urllib.request.urlopen(requests).read()
    soup = bs.BeautifulSoup(requestread,'html.parser')
    poem_content = (beautify_txt(soup.find_all('div', class_="poem_body")[0].text))
    poem_title = soup.find_all('h1')[0].text
    poems_results_RT.append([poem_title, poem_content])
  return poems_results_RT
```

## Pairwise similarity

Pairwise cosine similarity would just be the dot product of the vectors becasue tf-idf vectors from sklearn are already normalized and L2 norm of these vectors is 1. So denominator of cosine similiarity formula is 1 in this case.


```python
pairwise_similarities=np.dot(tfidf_vectors,tfidf_vectors.T)
pairwise_differences=euclidean_distances(tfidf_vectors)
```

```python
print (tfidf_vectors[0])
print (pairwise_similarities.shape)
print (pairwise_similarities[0][:])
```

similarity is highest, 1 at index 0 because they are the same documents

```python
def most_similar(doc_id,similarity_matrix,matrix):
    print (f'Poet: {data.iloc[doc_id]["Poet"]}')
    print (f'Poem: {data.iloc[doc_id]["Poem"]}')
    print ('\n')
    print (f'Similar Documents using {matrix}:')
    if matrix=='Cosine Similarity':
        similar_ix=np.argsort(similarity_matrix[doc_id])[::-1]
    elif matrix=='Euclidean Distance':
        similar_ix=np.argsort(similarity_matrix[doc_id])
    for ix in similar_ix:
        if ix==doc_id:
            continue
        print('\n')
        print (f'Poet: {data.iloc[doc_id]["Poet"]}')
        print (f'Poem: {data.iloc[ix]["Poem"]}')
        print (f'{matrix} : {similarity_matrix[doc_id][ix]}')
```

```python
# loading pre-trained embeddings, each word is represented as a 300 dimensional vector
import gensim

W2V_PATH="/content/drive/MyDrive/Google PreTrained/GoogleNews-vectors-negative300.bin.gz"
model_w2v = gensim.models.KeyedVectors.load_word2vec_format(W2V_PATH, binary=True)
```

```python
# creating embedding matrix, every row is a vector representation from the vocabulary indexed by the tokenizer index. 
embedding_matrix=np.zeros((vocab_size,300))
for word,i in tokenizer.word_index.items():
    if word in model_w2v:
        embedding_matrix[i]=model_w2v[word]
```

# Transpose verbs, Adjectives and nouns for both the Poets.

```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('stsb-roberta-large')
```

```python
def getSemanticSimilarity(p1,p2):
  embedding1 = model.encode(p1, convert_to_tensor=True)
  embedding2 = model.encode(p2, convert_to_tensor=True)
  cosine_scores=util.pytorch_cos_sim(embedding1,embedding2)
  similarity_score = cosine_scores.item()
  return similarity_score
```

### Swapping Verb based on similarity score. 

```python
def transposeVerb(p1,p2,text):
  for v1 in p1:
    for v2 in p2:
      score = getSemanticSimilarity(v1,v2)
      if score > 0.3:
        text  = text.replace(v1,v2)
        break
  return text     
```

### Swapping Adjectives based on similarity score. 

```python
def transposeAdjective(p1,p2,text):
  for a1 in p1:
    for a2 in p2:
      score = getSemanticSimilarity(a1,a2)
      if score > 0.3:
        text  = text.replace(a1,a2)
        break
  return text     
```

### Swapping Nouns based on similarity score. 

```python
def transposeNoun(p1,p2,text):
  for n1 in p1:
    for n2 in p2:
      score = getSemanticSimilarity(n1,n2)
      if score > 0.3:
        text  = text.replace(n1,n2)
        break
  return text     
```

### Swapping the Poems

```python
def transposePoem(poet1, poet2, text):
  p1_verbs = poet1['all_verbs'][0]
  p2_verbs = poet2['all_verbs'][0]
  print(p1_verbs)
  print(p2_verbs)
  new_poem = transposeVerb(p1_verbs,p2_verbs,text)
  
  print("new poem verbs")
  print(new_poem)
  
  p1_adj = poet1['all_adjectives'][0]
  p2_adj = poet2['all_adjectives'][0]
  print(p1_adj)
  print(p2_adj)
  new_poem1 = transposeAdjective(p1_verbs,p2_verbs,new_poem)
  print("new poem adjectives")
  print(new_poem1)
  
  p1_nn = poet1['all_nouns'][0]
  p2_nn = poet2['all_nouns'][0]
  print(p1_adj)
  print(p2_adj)
  final_new_poem = transposeAdjective(p1_verbs,p2_verbs,new_poem1)
  print("new poem adjectives")
  print(final_new_poem)
  return final_new_poem
```
